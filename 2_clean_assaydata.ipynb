### Load Library

```python
import pandas as pd
```

### Load Data

```python
input_file = "data/archive/a_OSD-944_metagenomic-sequencing_short-read-low-biomass-dna-sequencing_Illumina.txt"
output_file = "data/clean_assaydata.tsv"
```

```python
print("Loading raw assaydata...")
df = pd.read_csv(input_file, sep='\t')
print(df.shape)

# print(df.columns)
# df
# df.head()
```

### Copy Data

```python
# We want DNA concentration (QA Score), Sequencing Depth, and Contamination %
cols_to_keep = [
    'Sample Name',
    'Parameter Value[QA Score]',
    'Parameter Value[Read Depth]',
    'Parameter Value[Human Reads Removed]',
    'Parameter Value[Contaminant Reads Removed]'
]

clean_df = df[cols_to_keep].copy()

clean_df.columns = [
    'Sample_ID', 
    'DNA_Concentration_ng_ul', 
    'Total_Reads', 
    'Human_Reads_Removed_Pct', 
    'Blank_Contaminant_Removed_Pct'
]

clean_df['Sample_ID'] = clean_df['Sample_ID'].replace('Zymo_MMC_Log_11ng', 'Zymo_LD_MMC')

print(clean_df.shape)
clean_df
```

### Save

```python
clean_df.to_csv(output_file, sep='\t', index=False)
```
