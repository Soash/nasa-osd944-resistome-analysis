### Load Library

```python
import os
import requests
import pandas as pd
```

#### Function: download_resumable

```python
def download_resumable(url, dest_path):
    # Downloads a file from a URL, resuming if the file is partially downloaded.
    try:
        response = requests.head(url, timeout=10)
        response.raise_for_status()
        remote_size = int(response.headers.get('content-length', 0))
    except Exception as e:
        print(f"Failed to fetch info for {url}\nError: {e}")
        return

    local_size = 0
    if os.path.exists(dest_path):
        local_size = os.path.getsize(dest_path)

    if local_size == remote_size:
        print(f"Already fully downloaded: {os.path.basename(dest_path)}")
        return
    elif local_size > remote_size:
        print(f"Local file larger than remote. Restarting {os.path.basename(dest_path)}...")
        local_size = 0
        mode = 'wb'
    else:
        if local_size > 0:
            print(f"Resuming {os.path.basename(dest_path)} from byte {local_size}...")
        else:
            print(f"Starting {os.path.basename(dest_path)}...")
        mode = 'ab'

    headers = {'Range': f'bytes={local_size}-'} if local_size > 0 else {}

    try:
        with requests.get(url, headers=headers, stream=True, timeout=15) as r:
            r.raise_for_status()
            with open(dest_path, mode) as f:
                for chunk in r.iter_content(chunk_size=8192):
                    if chunk:
                        f.write(chunk)
    except Exception as e:
        print(f"Error downloading {url}\nError: {e}")
```

### Setup

```python
base_url = "http://nasa-osdr.s3-website-us-west-2.amazonaws.com/OSD-944/version-1/metagenomics-lowbiomass-shortread/"
metadata_file = "data/clean_metadata.tsv"
output_dir = "data/annotations"

os.makedirs(output_dir, exist_ok=True)
```

```python
print(f"Reading sample IDs from {metadata_file}...")

df = pd.read_csv(metadata_file, sep='\t')
target_samples = df['Sample_ID'].tolist()
print(f"Found {len(target_samples)} samples in metadata. Starting annotation downloads...")
```

### Download

```python
for sample in target_samples:
    filename = f"GLDS-775_metagenomics-lowbiomass-shortread_Prep3_{sample}-gene-coverage-annotation-and-tax.tsv"
    url = base_url + filename
    dest_path = os.path.join(output_dir, filename)
    
    download_resumable(url, dest_path)
```
